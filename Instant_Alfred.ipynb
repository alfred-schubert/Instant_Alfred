Import knihoven:

```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
import numpy as np
import io
from collections import Counter
```

```python
default_data = pd.read_excel("Test_data/Elektromobily.xlsx",header=[0,1])
```

Nahraj soubor:

```python
uploader = widgets.FileUpload(accept='.xlsx', multiple=False)
display(uploader)
```

Zpracování souboru:

```python
if uploader.value == ():
    full_data = default_data

else:
    loaded_file = uploader.value[0]
    full_data = pd.read_excel(io.BytesIO(uploaded_file.content), header=[0,1])

full_data.set_index(full_data.columns[0], inplace=True)
full_data.index.name = None
```

```python
quest_list = full_data.columns.get_level_values(0)
options_count = list(Counter(quest_list).items())
questions = pd.DataFrame(options_count, columns=["Otazky", "Odpovedi"])
questions.set_index(questions.columns[0], inplace=True)
questions['Volna otazka'] = False
```

Nyní vyber volné otázky (ctrl+click umožňuje vybrat více otázek):

```python
choose_open_q = widgets.SelectMultiple(
    options=questions.index,
    value=[],
    rows=questions.shape[0],
    description='Volné otázky',
    disabled=False
)
display(choose_open_q)
```

```python
questions['Volna otazka'] = False
questions.loc[list(choose_open_q.value), 'Volna otazka'] = True
questions['Typ otazky'] = None
questions.loc[questions['Volna otazka'] == True, 'Typ otazky'] = 'Volna'
questions.loc[(questions['Volna otazka'] == False) & (questions['Odpovedi'] > 1), 'Typ otazky'] = "Jednoducha"
questions.loc[(questions['Volna otazka'] == False) & (questions['Odpovedi'] == 1), 'Typ otazky'] = "Slozita"
```

```python
open_questions = questions.loc[questions['Volna otazka']==True].index
df = full_data.drop(columns=open_questions, level=0, axis=1)
```
